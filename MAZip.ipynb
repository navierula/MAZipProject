```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("MAZip.csv")
df.head(10)

df = df.iloc[:3849,[0,1,2,3,4,8]]

df = df.dropna() #dropped null vals

df = df[df["Number of returns"] != "**"]
df = df[df["Number of joint returns"] != "**"]
df = df[df["Number of dependents"] != "**"]

        
```

```python
df
```

```python
result = []
for i in df.index: 
    returns = df['Number of returns']
    dependents = df['Number of dependents']
for k, j in zip(dependents, returns):
    if k != "**" and j != "**":
        result.append((float(k)/(float(j) + 0.0001)))
print result[0:10]
print len(result)
#print result
```

```python
jr_list = []
for i in df.index: 
    joint_returns = df['Number of joint returns']
for j in joint_returns:
    if j != "**":
        jr_list.append(float(j))
    #print result[0:10]
print len(jr_list)
print jr_list[0:10]
```

```python
plt.scatter(result,jr_list)
plt.show()
```
